```python
from transformers import AutoTokenizer

model_name = "distilbert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(model_name)

print("Tokenizer Loaded Successfully")
```

```python
sample_text = train_df.iloc[0]["text"]

tokens = tokenizer(sample_text)

print(tokens.keys())
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(
    "distilbert-base-uncased",
    num_labels=2
)

print("Model Loaded Successfully")
```

```python
print(model.config)
```

```python
def tokenize_function(examples):
    return tokenizer(
        examples["text"],
        truncation=True,
        padding="max_length",
        max_length=256
    )
```

```python
tokenized_dataset = dataset.map(
    tokenize_function,
    batched=True
)
```

```python
tokenized_dataset["train"][0]
```

```python
import torch

print(torch.cuda.is_available())
print(torch.cuda.get_device_name(0))
```

```python
from kaggle_secrets import UserSecretsClient

secrets = UserSecretsClient()

print("HF Token Exists:", secrets.get_secret("HF_TOKEN") is not None)
print("W&B Token Exists:", secrets.get_secret("WANDB_API_KEY") is not None)
```

```python
small_train = tokenized_dataset["train"].shuffle(seed=42).select(range(5000))
small_test = tokenized_dataset["test"].shuffle(seed=42).select(range(1000))

print(len(small_train))
print(len(small_test))
```

```python
from huggingface_hub import login
import wandb
import os

HF_TOKEN = secrets.get_secret("HF_TOKEN")
WANDB_API_KEY = secrets.get_secret("WANDB_API_KEY")

login(token=HF_TOKEN)

os.environ["WANDB_API_KEY"] = WANDB_API_KEY

wandb.login()

print("Login Successful")
```

```python
wandb.init(
    project="imdb-distilbert",
    name="run-v1",
    config={
        "model":"distilbert-base-uncased",
        "epochs":1,
        "batch_size":16,
        "learning_rate":2e-5,
        "dataset":"IMDb",
        "version":"v1"
    }
)
```

```python
from sklearn.metrics import accuracy_score, f1_score

def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)

    return {
        "accuracy": accuracy_score(labels, preds),
        "f1": f1_score(labels, preds)
    }
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=1,
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    report_to="wandb",
    run_name="run-v1",
    logging_steps=100
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=small_train,
    eval_dataset=small_test,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

```python
wandb.finish()
```

```python
train_subset = tokenized_dataset["train"].shuffle(seed=42).select(range(10000))
test_subset = tokenized_dataset["test"].shuffle(seed=42).select(range(2000))

print(len(train_subset))
print(len(test_subset))
```

```python
wandb.init(
    project="imdb-distilbert",
    name="run-v3-final",
    config={
        "model":"distilbert-base-uncased",
        "learning_rate":5e-5,
        "epochs":3,
        "train_samples":10000,
        "test_samples":2000,
        "version":"final"
    }
)
```

```python
from transformers import AutoModelForSequenceClassification

model_final = AutoModelForSequenceClassification.from_pretrained(
    "distilbert-base-uncased",
    num_labels=2
)
```

```python
training_args_final = TrainingArguments(
    output_dir="./final_model",
    num_train_epochs=3,
    learning_rate=5e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    report_to="wandb",
    run_name="run-v3-final",
    logging_steps=100
)
```

```python
trainer_final = Trainer(
    model=model_final,
    args=training_args_final,
    train_dataset=train_subset,
    eval_dataset=test_subset,
    compute_metrics=compute_metrics
)
```

```python
trainer_final.train()
```

```python
wandb.finish()
```
